```python
import numpy as np 
from ipywidgets import interact
```

```python
x = 1.5
a = 2

x = (x+a/x)/2
print(x)
```

```python
x = (x+a/x)/2
print(x)
```

```python
x = (x+a/x)/2
print(x)
```

```python
def iterate_sqrt(a,n,x):
    for i in range(n):
        x = (x+a/x)/2
    return x 

iterate_sqrt(2,10,1) - np.sqrt(2)
```

```python
@interact(n=(0, 10))
def show_iteration(n=0):
    result = iterate_sqrt(2, n, 1)
    print(f"After {n} iterations: {result}")

```

```python
import matplotlib.pyplot as plt
# Using range(11) to match your n=(0, 10) slider
n_values = range(7)
# Calculating results (casting to float just in case you are using Decimal)
results = [float(iterate_sqrt(2, n, 1)) for n in n_values]
# Calculating absolute error vs the numpy reference
errors = [abs(res - np.sqrt(2)) for res in results]

log_errors = -1*np.log(errors)

rate = np.log(log_errors)

plt.figure(figsize=(12, 5))
# Subplot 1: Convergence of the value
plt.subplot(1, 2, 1)
plt.plot(n_values, results, 'o-', label='Iteration Value')
plt.axhline(y=np.sqrt(2), color='r', linestyle='--', label='Actual sqrt(2)')
plt.title('Convergence to sqrt(2)')
plt.xlabel('n (Iterations)')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
# Subplot 2: Convergence of the error (Log Scale)
plt.subplot(1, 2, 2)
plt.semilogy(n_values, errors, 'o-r')
plt.title('Error Magnitude (Log Scale)')
plt.xlabel('n (Iterations)')
plt.ylabel('Absolute Error')
plt.grid(True, which="both", ls="-")
plt.tight_layout()
plt.show()

plt.subplot(1, 3, 3)
plt.plot(n_values, rate, 'o-r')
plt.title('Rate')
plt.xlabel('n (Iterations)')
plt.ylabel('Rate')
plt.grid(True, which="both", ls="-")
plt.tight_layout()
plt.show()
```
